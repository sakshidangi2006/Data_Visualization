```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Matplotlib Basics

```python
x = np.linspace(0,5,11)
y = x**2
x , y
```

```python
plt.plot(x,y)
plt.title("Graph")
plt.xlabel("X axis")
plt.ylabel("Y axis")
```

Subplots in Matplotlib

```python
plt.subplot(2,2,1)
plt.plot(y,x)
plt.subplot(2,2,3)
plt.plot(y**2,x)
plt.subplot(2,2,4)
plt.plot(x**2,x**2)
plt.subplot(2,2,2)
plt.plot(x,y)
```

Object Oriented Way

Matplotlib allows the aspect ratio, DPI and figure size to be specified when 
the figure object is created. You can use the figsize and dip keyword arguments.
figsize is a tuple of the  width and height of the figure in inches and dpi is the dots-per-inch(pixel per inch)   

```python
fig = plt.figure(figsize = (10,8), dpi = 100)
axis1 = fig.add_axes([0.1,0.1,0.8,0.8])
axis1.plot(x,y)
axis2 = fig.add_axes([0.2,0.4,0.4,0.4])
axis2.plot(x,y)
```
