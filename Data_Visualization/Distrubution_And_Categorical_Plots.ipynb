Distribution Plot

We use distribution plot to visualize the distribution of quantative data

Some Plots that allow us to visualize the distribution of a data set. These plots are:
a. Distplot/histplot
b. jointplot
c. pairplot
d. rugplot

```python
df = sns.load_dataset('tips')
df
```

```python
df['size'].unique()
```

```python
plt.subplot(1,2,1)
sns.histplot(df['total_bill'],bins = 20, kde = True)
plt.subplot(1,2,2)
sns.histplot(df['tip'],bins = 20, kde = True)
```

Categorial Data Plots

There are a few main plot types for this

a. boxplot
b. violinplot
c. stripplot
d. swarmplot
e. barplot
f. countplot


Countplot

```python
sns.countplot(x = df['sex'], hue = df['smoker'])
```

BarPlot

```python
sns.barplot(x = df['sex'], y = df['total_bill'])
```

You can the estimator object to your own function , that converts a vector to a scaler 
Mean(default): np.mean
Median: np.median
Sum: np.sum
Count: len(to count the number of observations)
Standard Deviation: np.std
Minimum: np.min
Maximum: np.max

```python
sns.barplot(x = df['sex'], y = df['tip'] ,estimator = np.sum)
```

BoxPlot

```python
sns.boxplot(x = df['tip'] , y =df['day'],  palette='rainbow')
```

boxplot and violin boxplot and violinplots are used to shown the distribution of categorical data. A box plot (a box-and-whisker plot) shows the distribution of quantitative data in a way that facilitates comparison between variables or across levels of categorical variables. The box shows the quartiles of the datasets while the whiskers extend to show the rest of the distribution, except for points that are determined to be 'outliers' using a method that is a function of the inter-quertile range.

```python
sns.violinplot(x = 'day', y ='total_bill', data = df)
```

Stripplot and Swarmplot

The stripplot will draw a scatterplot where one variable is categorical . A strip plot can be drawn on its own, but it is also good complement to be a box or violin plot in case where ;you want to show all observations along with some representation of the underlying distribution .

The swarmplot is similar to stripplot(), but the points are adjusted (only along the categorical axis) so that they don't overlap. This gives a better representation of the distribution of values, although it does not scale as well to large numbers of observations (both in terms of the ability to show all the points and in terms of the computation needed to arrange them).

```python
sns.stripplot(x = 'day', y = 'total_bill', data = df)
```

```python
sns.swarmplot(x = 'day', y = 'total_bill', data = df)
```

```python
sns.violinplot(x='tip', y='day', data =df)
sns.swarmplot(x='tip', y='day', data =df)
```
